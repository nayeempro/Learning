```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df]) #add by row
df
```

```python
df = pd.concat([temperature_df,windspeed_df], axis = 1) #add by column
df
```

# Basic marge Using DataFrame

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df3 = pd.merge(df1,df2,on='city') #inner join (intersection)
df3
```

```python
df3 = pd.merge(df1,df2,how='outer') #outer join join (union)
df3 #inner by defult
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","san francisco"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city') #inner join (intersection)
df3 #just mille nibe
```

```python
df3 = pd.merge(df1,df2,how='outer') #outer join (union)
df3 #all nibe
```

```python
df3 = pd.merge(df1,df2,how='left') #outer join (union)
df3 #df1 all asbe
```

```python
df3 = pd.merge(df1,df2,how='right') #outer join (union)
df3 #df2 all asbe
```

```python
df3 = pd.merge(df1,df2,how='right',indicator=True) #outer join (union)
df3 #df2 all asbe
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```
