# transform or reshape data (melt function)

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/11_melt/weather.csv")
df
```

```python
df1 = pd.melt(df,id_vars=['day'])
df1
```

```python
df1[df1['variable']=='chicago']
```

```python
df1 = pd.melt(df,id_vars=['day'],var_name='city',value_name='temperature')
df1
```

```python
df1[df1['city']=='chicago']
```

# crosstable

```python
df = pd.read_excel('survey.xls')
df
```

```python
df1 = pd.crosstab(df.Nationality, df.Handedness)
df1
```

```python
df
```

```python
df1 = pd.crosstab(df.Sex, df.Handedness)
df1 #how many female are write handed = 3
```

```python
df1 = pd.crosstab(df.Nationality, df.Handedness,margins=True)
df1
```

```python
df1 = pd.crosstab([df.Nationality,df.Sex], [df.Handedness],margins=True)
df1 #how many female in USa are right handed =2
```

```python
df1 = pd.crosstab(df.Sex, df.Handedness,margins=True)
df1
```

```python
df1 = pd.crosstab(df.Sex, df.Handedness,normalize='index')
df1
```

```python
import numpy as np
df1 = pd.crosstab(df.Sex, df.Handedness,values=df.Age, aggfunc= np.average)
df1
```
